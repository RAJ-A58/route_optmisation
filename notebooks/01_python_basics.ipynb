```python
numbers=[1,2,3,4]
print(numbers)
#A Python **list** is a fundamental and highly versatile data structure characterized primarily by its
# ordered, mutable, and heterogeneous nature. Being **ordered** means that the items have a defined 
# sequence, and new items are generally added to the end, maintaining a stable indexing system starting 
# from 0. Crucially, lists are **mutable**, meaning they can be changed after creation—elements can be 
# added using methods like `.append()` or `.insert()`, removed using methods like `.pop()` or `.remove()`, 
# and modified by assigning a new value to a specific index. Lists are also **heterogeneous**; they can 
# contain elements of different data types, such as integers, strings, booleans, and even other lists, all
# within the same structure, which is visually delineated by square brackets `[]`.
print(numbers[0])
print(numbers[1])       
```

```python
for i in numbers:
    print(i*2)
```

```python
location={"id":1,"x":(10,20),"y":20}
print(location["id"])
print(location["x"])
print(location["y"])
print(location)
print(location.keys())
```

```python
def distance(a, b):  
    return abs(a - b)

print(distance(10, 3))
#functions implementation
```

```python
import numpy as np

coords = np.array([[0, 0], [3, 4]])
dist = np.linalg.norm(coords[0] - coords[1])
print(dist)
```
